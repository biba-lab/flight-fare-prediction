## 🧼 Pulizia dei dati

## 🕒 Creazione delle feature temporali
Aggiungiamo:
- Mese del viaggio
- Giorno della settimana

```python
df = aggiungi_feature_temporali(df)
print("✔️ Feature temporali aggiunte.")
df.head()
```

```python
df = aggiungi_durata_minuti(df)
print("✔️ Durata convertita in minuti.")
df[["Duration", "Duration_min"]].head()
```
